# Generate and plot 6 distributions

### Making a dictionary to host the distributions.

The dictionary includes six distributions for Normal, Poisson, Binomial, Chi-square, Exponential and Wald.

```python
from __future__ import print_function
__author__= 'yxz'
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as pl
import json
s = json.load(open('fbb_matplotlibrc.json'))
pl.rcParams.update(s)
md={}
means = {}
mymean = 100
mysize=2000/(np.array(range(1,100)+[10]))
```

### Generate and Plot

```python
# Generating 100 different Normal Distributions with SAME sample mean and DIFFERENT sample size
# Parameters for Normal Distribution:
# np.random.normal(mean, sigma, size)
np.random.seed(2015)
md['normal']={}
mu=mymean
sigma=1
for n in mysize:
    md['normal'][n] = np.random.normal(mu,sigma,n)
means['normal'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['normal'].iterkeys():
    means['normal'][nn] = md['normal'][nn].mean()
    axchisq_mu_n.plot(nn, means['normal'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')
```

```python
# Generating 100 different Poisson Distributions with SAME sample mean and DIFFERENT sample size
# Parameters for Poisson distribution:
# np.random.poisson(lambda, size)
# and the lambda=mean
np.random.seed(2015)
md['poisson']={}
ld=mymean
for n in mysize:
    md['poisson'][n] = np.random.poisson(ld,n)
    
means['poisson'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['poisson'].iterkeys():
    means['poisson'][nn] = md['poisson'][nn].mean()
    axchisq_mu_n.plot(nn, means['poisson'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k') 
```

```python
# Generating 100 different Binomial Distributions with SAME sample mean and DIFFERENT sample size
# Parameters for Binominal Distribution:
# np.random.binominal(n,p,size)
# and n*p=mean
# and 0<p<1
md['binomial']={}
myp=0.5
myn=mymean/myp
for n in mysize:
    md['binomial'][n] = np.random.binomial(myn,myp,n)

means['binomial'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['binomial'].iterkeys():
    means['binomial'][nn] = md['binomial'][nn].mean()
    axchisq_mu_n.plot(nn, means['binomial'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k') 
```

```python
# Generating 100 different Chisquare Distributions with SAME sample mean and DIFFERENT sample size
# Parameters for Chisquare Distribution:
# np.random.chisquare(df,size)
# and df=mean
md['chisq']={} 
df = mymean
for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size=n)
    
means['chisq'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['chisq'].iterkeys():
    means['chisq'][nn] = md['chisq'][nn].mean()
    axchisq_mu_n.plot(nn, means['chisq'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')
```

```python
# Generating 100 different Exponential Distributions with SAME sample mean and DIFFERENT sample size
# Parameters for Exponential Distribution:
# np.random.exponential(bt,size)
# and beta=mean
md['exp']={} 
bt = mymean
for n in mysize:
    md['exp'][n] = np.random.exponential(bt, size=n)
    
means['exp'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['exp'].iterkeys():
    means['exp'][nn] = md['exp'][nn].mean()
    axchisq_mu_n.plot(nn, means['exp'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')
```

```python
# Generating 100 different Wald Distributions with SAME sample mean and DIFFERENT sample size
# Parameters for Wald Distribution:
# np.random.wald(mean,scale,size)
md['wald']={}
mu=mymean
scale=2
for n in mysize:
    md['wald'][n] = np.random.wald(mu,scale,n)
    
means['wald'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['wald'].iterkeys():
    means['wald'][nn] = md['wald'][nn].mean()
    axchisq_mu_n.plot(nn, means['wald'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')
```

### From the six different distribution of means we can see, the larger the sample size, the closer is the sample mean to the population mean.

# Plot the distributions of all sample means

### Histogram of all sample means

```python
allmeans=[]
for d in means:
    for n in means[d]:
        allmeans.append(means[d][n])
pl.figure(figsize=(10,10))
pl.hist(allmeans, normed=1,alpha=0.5, bins=30)
pl.xlabel('sample mean')
pl.ylabel('N')
```

### Optional: Scatter plot, the same distribution with same color

```python
fig = pl.figure(figsize=(20,20))
axchisq_mu_n = fig.add_subplot(111)


axchisq_mu_n.set_xlabel('sample size')
axchisq_mu_n.set_ylabel('sample mean')
axchisq_mu_n.set_ylim(0,200)
axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')

colormap=['b', 'm', 'r', 'c', 'g', 'k']
for d,i in zip(means, range(len(means))):
    f=True
    for n in means[d]:
        if f:
            axchisq_mu_n.plot(n, means[d][n], '%s.'%colormap[i], label='%s'%d)
            f=False
        else:
            axchisq_mu_n.plot(n, means[d][n], '%s.'%colormap[i])
        axchisq_mu_n.hold(True)
axchisq_mu_n.legend()
```

### Optional: Fit a Gaussian to the distribution of means

```python
from scipy.stats import norm
import matplotlib.mlab as mlab

(mu, sigma) = norm.fit(allmeans)
pl.figure(figsize=(10,10))
n, bins, patches=pl.hist(allmeans,normed=1,bins=50, alpha=0.3)
y = mlab.normpdf( bins, mu, sigma)
l = pl.plot(bins, y, 'r', linewidth=2)
```
